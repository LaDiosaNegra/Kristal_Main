```python
df = pd.DataFrame({
    'grps': list('aaabbcaabcccbbc'),  # Group identifiers are the same
    'vals': [10, 300, 5, 2, 50, 10, 5, 50, 55, 25, 200, 20, 60, 4, 90]
})

df
```

```python
df.groupby('grps')['vals'].sum()
```

```python
df.groupby('grps')['vals'].nlargest(3)
```

```python
x = df.groupby('grps')['vals'].nlargest(5)
x
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.RandomState(4711).randint(1,51,size = (50,2)),columns = ['A','B'])
df
```

```python
pd.cut(df.A, np.arange(0,51, 5)) # groups the values in the column A into intervals of 5
```

```python
df.groupby(pd.cut(df.A, np.arange(0, 51, 10)))['B'].sum() # groups the values in column 'A' into bins of 10 then calculates the sum of corresponding values in column 'B' for each group
```

```python
df.A[5:15].sum() + df.B[5:15].sum() # calculates the sum of index 0 to 8 in columns 'A' and 'B'  then adds these two sums together
```

```python
df.groupby(pd.cut(df.A, np.arange(0, 51, 5)))['B'].nlargest(5)  #groups by intervals of 5 units in column 'A'and then for each group, then selects the five largest values from column 'B
```
