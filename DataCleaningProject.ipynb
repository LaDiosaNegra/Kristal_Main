```python
import pandas as pd
import numpy as np

# Creating a DataFrame with inconsistencies and some data issues
df= pd.DataFrame({
    'Name': ['ALIce_MarTin', 'bOB_sMITH', 'CHarlie_BrowN', 'David_joHNsTon', 'Eve_STONE'],
    'Age': [28, np.nan, 35, 42, np.nan],
    'Email': ['alice.martin@EXAMPLE.com', 'bob.smith@example.COM', 'charlie.brown@EXAMPLE.com', 'david.johnston@EXAMPLE.com', 'eve.stone@EXAMPLE.com'],
    'PhoneNumber': ['321-654-9870', '321.654.9871', '(321) 654-9872', '321 654 9873', '3216549874'],
    'Salary': [45000, 56000, np.nan, 64000, 76000],
    'Department': ['Marketing', 'Sales', 'Marketing', 'Operations', 'HR'],
    'DateOfJoining': ['2019-05-20', '2018-02-18', '2021-09-09', '2019-12-01', '2018-03-15'],
    'RecentProjects': [['Project X', 'Project Y'], ['Project Z'], [], ['Project W', 'Project V'], ['Project U', 'Project T']],
    'Manager': ['Rob', 'Mia', 'Nate', 'Oscar', 'Lia']
})

# Display the DataFrame
df
```

```python
df.dtypes
```

```python
y = df['Name'].str.split( "_", expand = True)
y
```

```python
y.columns = ['First Name', 'Last Name']
y
```

```python
y['First Name'] = y['First Name'].str.capitalize()
y
```

```python
y['Last Name'] = y['Last Name'].str.capitalize()
y
```

```python
df = df.join(y)
df
```

```python
df = df.drop('Name', axis = 1)
df
```

```python
df['Email'] = df['Email'].str.lower()
df
```

```python
df2 = df.copy()
df2
```

```python
df2['DateOfJoining'] = pd.to_datetime(df2.DateOfJoining)
df2
```

```python
df2['PhoneNumber'] = df2['PhoneNumber'].str.replace(r'\D','', regex = True)
df2
```

```python
y = df.RecentProjects.apply(pd.Series)
y
```

```python
y.columns = ['Project_1', 'Project_2']
y
```

```python
sample = []
string = 'Project_'

for i in range(1, len(y.columns) + 1):
  sample.append(string + str(i))

sample


```

```python
y.columns = sample
y
```

```python
df2 = df2.join(y)
df2
```

```python
df2 = df2.drop('RecentProjects', axis = 1)
df2
```

```python
df2['Age']  = df2['Age'].astype(int)
df2
```

```python
df2['Age'].fillna(df2['Age'].median(), inplace = True)

df2
```

```python
df2['Age'] = df2['Age'].astype(int)
df2
```

```python
df2['Salary'] = df2['Salary'].interpolate()
df2
```

```python
df2['Salary'] = df2['Salary'].astype(int)
df2
```

```python
df2['Project_1'].fillna('No Project Assigned', inplace = True)
df2['Project_2'].fillna('No Project Assigned', inplace = True)
df2
```

```python

df2 = df2[['First Name', 'Last Name', 'Email', 'Age', 'PhoneNumber', 'DateOfJoining', 'Salary', 'Department', 'Project_1', 'Project_2', 'Manager']]
df2
```
