```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'Name': ['JoHn_doe', 'JANE_SmiTH', 'miChael_browN', 'kATE_wilSON', 'MAtt_heWson'],
    'Age': [25, 30, np.nan, 45, np.nan],
    'Email': ['john.doe@example.com', 'jane.smith@EXAMPLE.com', 'michael.BROWN@example.com', 'kate.wilson@EXAMPLE.com', 'matt.hewson@EXAMPLE.com'],
    'PhoneNumber': ['123-456-7890', '123.456.7890', '(123) 456-7890', '123 456 7890', '1234567890'],
    'Salary': [50000, 60000, 55000, np.nan, 70000],
    'Department': ['Engineering', 'HR', 'Engineering', 'HR', 'Finance'],
    'DateOfJoining': ['2020-01-15', '2019-03-21', '2021-07-11', '2022-11-01', '2020-02-05'],
    'RecentProjects': [['Project A', 'Project B'], [], ['Project C', 'Project D', 'Project E'], ['Project F'], ['Project G', 'Project H']],
    'Manager': ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
})
```

```python
df
```

```python
df.dtypes
```

```python
z = df['Name'].str.split( "_", expand = True)
z
```

```python
z.columns = ['First Name', 'Last Name']
z
```

```python
z['First Name'] = z['First Name'].str.capitalize()
z
```

```python
z['Last Name'] = z['Last Name'].str.capitalize()
z
```

```python
df = df.join(z)
df
```

```python
df = df.drop(['Name'], axis = 1)
df
```

```python
df['Email'] = df['Email'].str.lower()
df
```

```python
data = df.copy()
data
```

```python
data['PhoneNumber'] = data['PhoneNumber'].str.replace(r'\D',"" ,regex = True)
data
```

```python
data['DateOfJoining'] = pd.to_datetime(data.DateOfJoining)
data
```

```python
data.dtypes
```

```python
max_projects = data['RecentProjects'].apply(len).max()
for i in range(max_projects):
    data[f'Project_{i+1}'] = data['RecentProjects'].apply(lambda x: x[i] if len(x) > i else np.nan)
```

```python
data
```

```python
data = data.drop(['RecentProjects'], axis = 1)
data
```

```python
data = data[['First Name', 'Last Name', 'Age' , 'Email', 'PhoneNumber', 'Salary', 'Department', 'DateOfJoining', 'Manager', 'Project_1', 'Project_2', 'Project_3']]
data
```
