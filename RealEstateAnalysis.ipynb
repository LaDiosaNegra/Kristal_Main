```python
# Import the regular expression library
import re

# Define a function to find and delete numbers in the 'Location' column

def find_and_delete_numbers(text):
    # Use regular expression to find all numbers and replace them with an empty string
    return re.sub(r'\d+', '', text)

# find_and_delete_numbers function to the 'Location' column
data['Location'] = data['Location'].apply(find_and_delete_numbers)

# Display the DataFrame after updating the 'Location' column
display(data.head(200))
```

```python


def remove_non_float(text):
    # Check if the value is a string and not missing
    if pd.notnull(text) and isinstance(text, str):
        # Use regular expression to find and keep only float values
        return re.sub(r'[^\d.]+', '', text)
    else:
        # If it's not a string or it's missing, return it as it is
        return text

# Example usage with a pandas DataFrame:
# Assume 'data' is your DataFrame and 'Price' is the column you're cleaning.
data['Price'] = data['Price'].apply(remove_non_float)

# Convert the 'Price' column to float type, invalid parsing will be set as NaN
data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

# Display the first few rows of the DataFrame to check the results
print(data)
```

```python

# Define a function to remove numbers from the 'House Type' column

def remove_numbers_from_house_type(text):
    # Use regular expression to find and remove all numbers
    return re.sub(r'\d+', '', text)

# Apply the function to the 'House Type' column
data['House Type'] = data['House Type'].apply(remove_numbers_from_house_type)

# Display the DataFrame after applying the function
display(data.head())
```

```python


def clean_house_type(text):
    # Remove any numbers
    text = re.sub(r'\d+', '', text)
    # Define the allowed house types
    allowed_types = ['apartment', 'furnished apartment', 'penthouse apartment']
    # Check if the text is one of the allowed types
    if text.lower().strip() in allowed_types:
        return text
    else:
        return 'Other'  # Replace with 'Other' if not an allowed type

# Apply the function to the 'House Type' column
data['House Type'] = data['House Type'].apply(clean_house_type)

# Display the DataFrame after applying the function
display(data.head())
```

```python


# Define a function to remove anything that is not a string from the 'Location' column

def clean_location(text):
    # Use regular expression to keep only alphabetic characters and spaces
    return re.sub(r'[^a-zA-Z\s]', '', text)

# Apply the function to the 'Location' column
data['Location'] = data['Location'].apply(clean_location)


display(data.head())
```

```python
# Cleaning the 'Location' and 'Price' Columns


# Define a function to remove anything that is not a string in the 'Location' column
def clean_location(text):
    # Use regular expression to keep only alphabetic characters and spaces
    return re.sub(r'[^a-zA-Z\s]', '', text)

# Apply the function to the 'Location' column
data['Location'] = data['Location'].apply(clean_location)

# Define a function to delete any row that does not have floats or integers as values in the 'Price' column
def is_float_or_int(value):
    try:
        float(value)
        return True
    except ValueError:
        return False

data = data[data['Price'].apply(is_float_or_int)]

display(data.head())
```

```python
# Adding a 'USD Price' Column and Converting 'Price' to USD

# Define the exchange rate
exchange_rate_ghs_to_usd = 0.84

# Define a function to convert prices from GHS to USD
def convert_to_usd(price_ghs):
    if pd.notna(price_ghs): #this line checks if the input price_ghs is not a missing or null value.
        return float(price_ghs) * exchange_rate_ghs_to_usd
    else:
        return None  # Handle non-numeric values

# Apply the function to the 'Price' column and create a new 'USD Price' column
data['USD Price'] = data['Price'].apply(convert_to_usd)


display(data.head())
```

```python
# Replacing 'penthouseapartment' with 'penthouse apartment' in the 'House Type' column

data['House Type'] = data['House Type'].str.replace('penthouseapartment', 'penthouse apartment', regex=False)


print('DataFrame after replacing values in House Type column:')
display(data.head())
```

```python

# Convert 'USD Price' to numeric
# use 'coerce' to turn invalid parsing into NaN
#  drop these NaN values for accurate calculations
data['USD Price'] = pd.to_numeric(data['USD Price'], errors='coerce').dropna()

# Calculate mean, median, and average
mean_usd_price = data['USD Price'].mean()
median_usd_price = data['USD Price'].median()
avg_usd_price = data['USD Price'].mean()  # Average is the same as mean

# Calculate outliers
# Outliers are typically defined as values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR
Q1 = data['USD Price'].quantile(0.25)
Q3 = data['USD Price'].quantile(0.75)
IQR = Q3 - Q1
lowest_outlier = Q1 - 1.5 * IQR
highest_outlier = Q3 + 1.5 * IQR

# Display the results
print(f'Mean USD Price: {mean_usd_price}')
print(f'Median USD Price: {median_usd_price}')
print(f'Average USD Price: {avg_usd_price}')
print(f'Lowest Outlier: {lowest_outlier}')
print(f'Highest Outlier: {highest_outlier}')
```
