```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('Microsoft Dataset.csv')

df.head()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
# convert date column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
df.Date
```

```python
duplicate_counts = df.duplicated().sum()
duplicate_counts
```

```python
price_columns = ['Open', 'High', 'Low', 'Close', 'Adj Close'] #creates a list of column names
plt.figure(figsize = (15, 10))

for i, column in enumerate(price_columns):  #iterates over each column name stored in 'price_columns'
  plt.subplot(3, 2, i + 1)    #divides the figure into a grid of plots (3 rows and 2 columns) and places the next plot in the position i+1
  sns.histplot(df[column])  #histogram for the current column of the dataset (df[column])
  plt.title(f' Distribution of {column}')
plt.tight_layout()
```

```python
plt.figure(figsize = (15, 10))
sns.lineplot(x = 'Date', y = 'Volume', data = df[:200])
plt.title('Volume over time')
plt.xticks(rotation = 45 )
plt.show()
```

```python
plt.figure(figsize = (15, 10))
sns.scatterplot(x = 'Open', y = 'Close', data = df[:200])
plt.title('Open vs Close Prices')
plt.show()
```
