```python
import pandas as pd
import numpy as np

# Create a dictionary of car information
car_dict = {}
car_dict["Names"] = ["Toyota Camry" , "Honda Accord", "Nissan Altima" , " Hyundai Sonata",
                    "Volkswagen Passat", "Prosche 911", "BMW M3", "Mercedes-AMG C63 S",
                    "Audi RS 5 Sportback", "Audi SQ5", "Genesis G90", "Jaguar XJ", "Honda Pilot",
                    "Mazda 3", "Kia Rio", "Nissan Maxima", "Kantanka Onantefo", "Tesla Model 3",
                    "Honda Civic", "Alfa Romeo Giulia", "Infiniti Q50", "Bugatti Veyron",
                    "Nissan Sentra", "Mazda3", "Audi A3", "BMW 1 Series", "Acura TLX",
                    "Mercedes-Benz A-Class", "Lexus IS", "Jaguar XF", "Volvo V60",
                    "Jeep Grand Cherokee", "Pagani Huayra", "Mitsubishi Outlander",
                    "Bugatti Veyron", "McLaren 720S","Land Rover Discovery", "Ford F-150",
                    "Lexus RX",'Kia Sportage','Renault Clio', 'Peugeot 3008', 'Suzuki Swift',
                    'Jeep Wrangler', 'Volkswagen Polo', 'Maserati Levante', 'Bugatti Chiron',
                    'Bentley Continental GT', 'Rolls-Royce Phantom','Ford Mustang']

# Generate random engine capacity and price
engine_capacity = np.random.randint(1, 6, 50)
price = np.random.normal(90000, 15000, 50)


# Create a dictionary of car information
car_dict["Engine Capacity"] = engine_capacity
car_dict["Price"] = price

# Creating a DataFrame
car_df = pd.DataFrame(car_dict)

# Adding dollar signs and commas to the prices, and only two decimals
price = ["{:,.2f}".format(p) for p in price]

print(car_df)

# Calculate the average cost of the cars
average_cost = car_df['Price'].mean()

# Find the maximum and minimum car prices
max_price = car_df['Price'].max()
min_price = car_df['Price'].min()

# Print the avg, min, and max
print(f"Average cost: {average_cost:,.2f}")
print(f"Maximum car price: {max_price:,.2f}")
print(f"Minimum car price: {min_price:,.2f}")
```
