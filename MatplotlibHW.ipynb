```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(0)      #Ensure ransom data is the same every time we run it
data = np.random.randint(1,100, size = 100)
colors = np.random.rand(100)
sizes = np.random.randint(50, 500, 100)  # Sizes for each point
```

```python
plt.figure(figsize = (10, 6))
scatter = plt.scatter(data, sizes, c = colors, cmap = 'RdYlGn_r', s = sizes)
plt.colorbar(scatter)
plt.title('Scatter Plot with Color Bar')
plt.xlabel('Data Value')
plt.ylabel('Sizes')
plt.show()
```

```python
plt.figure(figsize = (14, 10))

plt.subplot(2, 2, 1)                           # prepares the area for the first chart
plt.hist(data, bins = 50, color = 'blue')
plt.title('Histogram of Data')
plt.xlabel('Data Range')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.subplot(2, 2, 2)
plt.bar(['Size 1', 'Size 2', 'Size 3'], data [:3])  #creates a bar chart for the first three categories
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
plt.subplot(2, 2, 3)
plt.barh(['Size 1', 'Size 2', 'Size 3'], data[:3])
plt.title('Horizontal Bar Chart')
plt.xlabel('Values')
plt.ylabel('Category')
plt.show()
```

```python

sales = np.random.randint(100, 1000, size = 100)
profit_margin = np.random.randint(10, 100, size = 100)
regions = np.random.randint(1, 4, size = 100) # sales region coded as 1, 2, 3

plt.figure()
scatter = plt.scatter(sales, profit_margin, c = regions, cmap = 'viridis', s = profit_margin *10)
plt.colorbar()
plt.title('Sales Amount vs Profit Margin')
plt.xlabel('Sales Amount ($)')
plt.ylabel('Profit Margin (%)')
plt.show()
```

```python

days = np.arange(1, 101)
temperatures = np.random.randint(20, 35, size = 100)
Humidity = np.random.randint(30, 100, size = 100)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('big_mart_sales.csv')
data.head()
```

```python
data.fillna({'Item_Weight': data['Item_Weight'].mean(), 'Outlet_Size': 'Medium'}, inplace = True)
```

```python
data.head()
```

```python
plt.figure(figsize = (10,6))
sns.violinplot(x ='Outlet_Size', y = 'Item_Outlet_Sales', data = data)
plt.title('Distribution of Outlet Sales by Outlet Size')
plt.show()
```

```python
sns.lineplot(figsize = (15,10))
sns.lineplot(x = 'Item_Type', y = 'Item_Visibility', data = data, estimator = 'mean') #average visibility of each item type in the dataset
plt.xticks(rotation = 90)
plt.title('Average Item MRP by Fat Content')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(x = 'Item_Fat_Content', y = 'Item_MRP', data = data[:200] , estimator = np.mean, width = 0.25)
plt.title('Average Item MRP by Fat Content')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.barplot(x = 'Item_MRP', y = 'Item_Fat_Content', data = data[:200] , estimator = np.mean, width = 0.25)
plt.title('Average Item MRP by Fat Content')
plt.show()
```

```python
plt.figure(figsize = (10, 6))
sns.histplot(data = data, x = 'Item_MRP', bins = 30, color = 'lavender')
plt.title('Distribution of Item MRP')
plt.show()
```

```python
colors = sorted(np.random.randint(0, 100, 200))
sns.relplot(x = 'Item_MRP', y = 'Item_Outlet_Sales', hue = 'Outlet_Size', kind = 'scatter', data = data[:200])
plt.title('Item MRP vs Outlet Sales Colored by Outlet Size')
plt.show()
```

```python

sns.catplot(x = 'Outlet_Size', y = 'Item_Outlet_Sales', kind = 'swarm', data = data[:500], hue = 'Outlet_Size')
plt.title('Swarm Plot of Item Sales by Outlet Size')
plt.show()
```
