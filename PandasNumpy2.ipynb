```python
import pandas as pd
import numpy as np


z = pd.date_range(start = '1997-10-01', end = '2007-10-01', freq = 'B') #  creates a sequence business days (excluding weekends) from October 1, 1997, to October 1, 2007
z
```

```python
len(z)
```

```python
y = pd.Series(np.random.rand(len(z)), index = z) #creates a pandas Series named y with random values between 0 and 1, these values are assigned to each date in the sequence. The dates in z serve as the index for the series
y
```

```python
z = pd.DataFrame(y)
z
```

```python
z[z.index.dayofweek == 4] # filters the date range to only include the dates where the day of the week is 4, which corresponds to Fridays
```

```python
z[z.index.dayofweek == 4].sum()
```

```python
z.resample('M').mean() #resamples a time series indexed by the date range to a monthly frequency and calculates the mean for each month
```

```python
z.resample('Y').min()
```

```python
z.resample('W').max()
```

```python
z.resample('W').sum()
```

```python
pd.Grouper(freq = '3M')
```

```python
z.groupby(pd.Grouper(freq = '3M')).idxmax() #groups the date range into three-month intervals and finds the index of the maximum value within each group
```
