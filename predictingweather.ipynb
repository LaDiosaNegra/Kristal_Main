```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

weather = pd.read_csv('local_weather.csv', index_col = "DATE")

weather
```

```python
weather.index = pd.to_datetime(weather.index)
weather.loc['1960-01-01' : '1960-01-31', :]
```

```python
# checking for missing data
weather.isnull().sum()
```

```python
threshold = 8000

# dropping columns with missing value rate higher than threshold
weather = weather[weather.columns[weather.isnull().mean() < threshold / len(weather)]]

#print null values in each column
weather.isnull().sum()
```

```python
# fill the missing values in the 'PRCP', 'SNOW', 'SNWD', 'TMAX', and 'TMIN'
#columns using their respective median values

weather.fillna(weather.median(), inplace = True)

# recheck missing data
weather.isnull().sum()
```

What is the range of temperatures for the dataset?

```python
# identify range of temperatures

temperature_range = weather['TMAX'].max() - weather['TMIN'].min()
temperature_range
```

 What is the highest amount of precipitation recorded in a day?

```python
# highest amount of precipitation in a day
max_precipitation = weather['PRCP'].max()
max_precipitation
```

Which station has recorded the highest amount of snow?

```python
# finding station with maximum snowfall
max_snow_station = weather[weather['SNOW']== weather['SNOW'].max()]['STATION']

max_snow_station.iloc[0]
```

How does the maximum temperature vary across different weather stations?

```python
# group data by "station" and calculate the mean of 'tmax'
avg_max_temp = weather.groupby('STATION')['TMAX'].mean()

avg_max_temp.head()
```

How does the minimum temperature vary across different weather stations?


```python
# group data by "station" and calculate the mean of 'tmin'
avg_min_temp = weather.groupby('STATION')['TMIN'].mean()

avg_min_temp.head()
```

Is there any correlation between precipitation and temperature?

```python
# since 'TAVG' doesn't exist in our dataset, we can create it as an avg of 'TMAX' and 'TMIN'
weather['TAVG'] = (weather['TMAX'] + weather['TMIN']) / 2

# compute correlatin between 'TAVG' and 'PRCP'
correlation = weather[['TAVG', 'PRCP']].corr()

correlation

plt.figure(figsize = (10, 6))
plt.scatter(weather['TAVG'], weather['PRCP'], alpha = 0.5)
plt.xlabel('Average Temperature')
plt.ylabel('Precipitation')
plt.title('Precipitation vs Average Temperature')
plt.show()
```

How does the amount of snowfall change across different seasons?

```python

#create a new 'Month' column
weather['Month'] = weather.index.month

def get_season(month):
  if month >= 3 and month <= 5:
    return 'Spring'
  elif month >= 6 and month <= 8:
    return 'Summer'
  elif month >= 9 and month <= 11:
    return 'Autumn'
  else:
    return "Winter"

# apply function to 'Month' column
weather['Season'] = weather['Month'].apply(get_season)

#group by 'Season' and calculate the mean of "SNOW"
snow_by_season = weather.groupby('Season')['SNOW'].mean()

snow_by_season


#plot avg snowfall in each season
snow_by_season.plot(kind = 'bar', figsize = (10, 6))
plt.ylabel('Average Snowfall')
plt.title('Average Snowfall by Season')
plt.show()
```

Seasonal Patterns: Is there a cyclical pattern in precipitation levels throughout the year? Can we predict the amount of rainfall in the coming months based on historical patterns

```python
# calculate average precipitation for each month over all the years
grouped = weather.groupby('Month')['PRCP'].mean()

#plot avg precipitation
plt.figure(figsize = (10, 5))
plt.plot(grouped)
plt.xlabel('Month')
plt.ylabel('Average Precipitation')
plt.title('Average Monthly Precipitation')
plt.show()
```

Are there any outliers?



```python
#variables to check
variables = ['TMAX', 'TMIN', 'PRCP']

plt.figure(figsize = (10,8))

# for each variable, create a subplot and plot a boxplot
for i, var in enumerate(variables):
  plt.subplot(len(variables), 1, i + 1)
  sns.boxplot(data = weather[var])
  plt.title(f'Boxplot of {var}')

plt.tight_layout()
plt.show()
```

Here's what we can interpret from the boxplots:

- **TMAX (Max temperature):** Most values fall approximately between -10 and 40 degrees. There are quite a few outliers on the lower side, indicating days with extraordinarily low maximum temperatures. There are also a few outliers on the higher side, representing extraordinarily hot days.

- **TMIN (Min temperature):** The range is broader, approximately between -20 and 25 degrees. Just like 'TMAX', 'TMIN' has several outliers on both the lower and higher side, referring to exceptionally cold and warm nights, respectively.

- **PRCP (Precipitation):** Most of the days have precipitation close to zero, but there are many outliers indicating days with heavy rainfall or snowfall.
