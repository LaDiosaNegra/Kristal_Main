```python
import pandas as pd





# Miami Heat Team Stats
url = "https://www.espn.co.uk/nba/team/stats/_/name/mia/miami-heat"

# Scrape tables from URL
miami_heat_stats = pd.read_html(url)
#print(miami_heat_stats)


# Select the longest table
longest_table = miami_heat_stats[0]

#Display basic info (data types of each column, the number of non-null (non-missing) values, and the memory usage.)
print(longest_table.info())

# Summary Statistics for numeric columns (count, avg, std, min/max values)
print(longest_table.describe())

# Dimensions of row and columns (returns a tuple representing the number of rows and columns)
#print(longest_table.shape)

# total number of elements
#print(longest_table.size)

#  extracts the "FT" column; isolating just the data in the "FT" column for analysis: avg FT % per game
mode_FT = longest_table["FT%"].mode()
print(mode_FT)

# Save the data to a CSV file
miami_heat_stats.to_csv("miami_heat_stats.csv", index=False)









```

```python
import pandas as pd

# URL of the website
url = "https://en.wikipedia.org/wiki/Greater_Accra_Region"

# Scrape the tables from the website
african_uni_IT = pd.read_html(url)

# Get the longest table
longest_table = african_uni_IT[0]

# Print the longest table
print(longest_table)

#Display basic info (data types of each column, the number of non-null (non-missing) values, and the memory usage.)
print(longest_table.info())

# Summary Statistics for numeric columns (count, avg, std, min/max values)
print(longest_table.describe())

# Dimensions of row and columns (returns a tuple representing the number of rows and columns)
print(longest_table.shape)

# total number of elements
print(longest_table.size)

# Print the correlation of the dataframe
print(longest_table.corr())
```
